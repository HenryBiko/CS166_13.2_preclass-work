```python
import random
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import describe
```

```python
def coin_challenge(strategy, bias=.6, capital=250, n=20):
    for _ in range(n):
        capital += strategy(capital) * (1 if random.random() < bias else -1)
    return capital

def twentyPercent(money):
    return money * 0.2

def justten(money):
    return 10

def tenPercent(money):
    return money * 0.1

def fiftyPercent(money):
    return money * 0.5
```

```python
twenty = [coin_challenge(twentyPercent) for _ in range(10000)]
print(describe(np.array(twenty)))
plt.hist(twenty)
plt.title("Always bet 20%")
plt.show()

ten = [coin_challenge(tenPercent) for _ in range(10000)]
print(describe(np.array(ten)))
plt.hist(ten)
plt.title("Always bet 10%")
plt.show()

fifty = [coin_challenge(fiftyPercent) for _ in range(10000)]
print(describe(fifty))
plt.hist(fifty)
plt.title("Always bet 50%")
plt.show()
```
